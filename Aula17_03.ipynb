```python
def main():
    carrinho = {}
    precos = {
        "Leite": 5.00,
        "Fini": 13.00,
        "Ovo": 15.00,
        "Mortadela": 20.00
    }

    while True:
        print("Bem-Vindo ao Mercado")
        print("----------------")
        print("Escolha uma Opção:")
        print("1. Comprar")
        print("2. Ver carrinho")
        print("3. Tirar item do Carrinho")
        print("4. Pagar e ir Embora")
        print("----------------")

        opcao = input("Opção: ")

        if opcao == "1":
            while True:
                print("Qual setor você quer passar?")
                print("1. Hortifruti")
                print("2. Diversos")
                print("3. Voltar")
                print("----------------")

                setor = input("Setor: ")

                if setor == "2":
                    while True:
                        print("O que você quer adicionar ao seu carrinho?")
                        print("1. Leite - R$5,00")
                        print("2. Fini - R$13,00")
                        print("3. Ovo - R$15,00")
                        print("4. Mortadela 100g - R$20,00")
                        print("5. Voltar")
                        print("----------------")

                        produto_escolhido = input("Escolha um número: ")

                        produtos = {"1": "Leite", "2": "Fini", "3": "Ovo", "4": "Mortadela"}

                        if produto_escolhido == "5":
                            print("Voltando...")
                            break

                        if produto_escolhido in produtos:
                            produto = produtos[produto_escolhido]
                            quantidade = int(input(f"Quantas unidades de {produto} deseja adicionar? "))
                            carrinho[produto] = carrinho.get(produto, 0) + quantidade
                            print(f"{quantidade}x {produto} foi adicionado ao carrinho!")
                        else:
                            print("Opção inválida!")

                elif setor == "3":
                    break
                else:
                    print("Setor inválido!")

        elif opcao == "2":
            print("Seu carrinho:")
            total = 0

            if not carrinho:
                print("Carrinho vazio!")
            else:
                for item, quantidade in carrinho.items():
                    preco = precos[item] * quantidade
                    total += preco
                    print(f"{quantidade}x {item} - R${preco:.2f}")
                print(f"Total: R${total:.2f}")

        elif opcao == "3":
            item_remover = input("Qual item deseja remover? ")

            if item_remover in carrinho:
                del carrinho[item_remover]
                print(f"{item_remover} foi removido do carrinho!")
            else:
                print("Este item não está no carrinho!")

        elif opcao == "4":
            print("Finalizando compra...")
            total = sum(precos[item] * quantidade for item, quantidade in carrinho.items())
            print(f"Total da compra: R${total:.2f}")
            print("Obrigado por comprar! Volte sempre.")
            break

        else:
            print("Opção inválida!")

if __name__ == "__main__":
    main()
```

```python
class Tabela:
    # Códigos ANSI para cores de texto
    RESET = "\033[0m"
    VERMELHO = "\033[31m"
    VERDE = "\033[32m"
    AMARELO = "\033[33m"
    AZUL = "\033[34m"
    MAGENTA = "\033[35m"
    CIANO = "\033[36m"

    # Códigos ANSI para cores de fundo
    FONDO_VERMELHO = "\033[41m"
    FONDO_VERDE = "\033[42m"
    FONDO_AMARELO = "\033[43m"
    FONDO_AZUL = "\033[44m"
    FONDO_MAGENTA = "\033[45m"
    FONDO_CIANO = "\033[46m"
    FONDO_BRANCO = "\033[47m"

    @staticmethod
    def imprimir_tabela():
        # Cabeçalho da tabela
        cabecalho = ["ID", "Nome", "Idade", "Cargo"]

        # Dados da tabela
        dados = [
            ["1", "Thiago", "17", "Aprendiz"],
            ["2", "Maria", "25", "Analista"],
            ["3", "João", "30", "Gerente"]
        ]

        # Imprime o cabeçalho com cor e fundo
        print(f"{Tabela.FONDO_CIANO}{Tabela.CIANO}{cabecalho[0]:<10}{cabecalho[1]:<10}{cabecalho[2]:<10}{cabecalho[3]:<10}{Tabela.RESET}")

        # Imprime os dados com cores alternadas para cada linha e fundo colorido
        for i, linha in enumerate(dados):
            cor_linha = Tabela.FONDO_VERDE + Tabela.VERDE if i % 2 == 0 else Tabela.FONDO_AMARELO + Tabela.AMARELO
            print(f"{cor_linha}{linha[0]:<10}{linha[1]:<10}{linha[2]:<10}{linha[3]:<10}{Tabela.RESET}")

if __name__ == "__main__":
    Tabela.imprimir_tabela()
```
